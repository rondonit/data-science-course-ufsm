```python
# ao invés de dropar as linhas, é possível preencher com a média/mediana

# list comprehension para pegar apenas as colunas que tem NaN
colunas_com_nan = [col for col in dados.columns if dados[col].isnull().any()]

colunas_com_nan
```

```python
def conta_outliers(df):
    outliers_count = {}
    for col in df.columns:
        # definir a distancia interquartil IQR
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1

        # definir os limite para outliers
        limite_inferior = q1 - 1.5 * iqr
        limite_superior = q3 + 1.5 * iqr

        # contar os outliers para a coluna (feature) atual
        outliers = df[(df[col] < limite_inferior) | (df[col] > limite_superior)]
        outliers_count[col] = len(outliers)

    return outliers_count
```

```python
d = dados.copy()
del d['diagnosis']

c = conta_outliers(d)
outliers_count_ordenado = dict(sorted(c.items(), key=lambda item: item[1], reverse=True))

outliers_count_ordenado
```

# Amostragem Estratificada ⚗️🧪

```python
classes = dados['diagnosis'].unique()
tamanho_amostra = 10
qtde_por_classe = round(tamanho_amostra / len(classes))

amostra_por_classe = []

for c in classes:
    indices_c = dados['diagnosis'] == c
    obs_c = dados[indices_c]
    proporcao_c = len(obs_c) / len(dados)
    qtd_c = round(proporcao_c * tamanho_amostra)
    amostra_c = obs_c.sample(qtd_c)
    amostra_por_classe.append(amostra_c)

amostra_estratificada = pd.concat(amostra_por_classe)
amostra_estratificada
```

```python
proporcao_amostra_estratificada = amostra_estratificada['diagnosis'].value_counts(normalize=True)
proporcao_amostra_estratificada
```
